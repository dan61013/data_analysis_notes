```python
import pandas as pd
```

```python
# Load DataFrame
df = pd.read_csv("../input/學生修課資料與成績.csv",
                 encoding="utf-8-sig",
                 low_memory=False)
```

```python
df.head()
```

`pandas.pivot()`

parameter:
1. data: DataFrame
2. index: index column, 可用list, array指定多層
3. values: 篩選數據，可用list, array指定多條件
4. aggfunc: 自定義方法

```python
student_class = df.pivot_table(
    index=["就讀班級"],
    aggfunc={"成績": ["mean", "count"]}
)
student_class
```
