```python
import pandas as pd
```

```python
df = pd.read_csv(
    '../dataset/學生修課資料與成績.csv',
    encoding='utf-8-sig'
)
```

```python
df.head()
```

```python
filter_mask = (df['成績'] >= 60)
```

```python
df[filter_mask].head()
```

```python
df.loc[(df['成績'] >= 60)].head()  # mask的用法，直接使用[]，或使用loc的方法相同
```

```python
# multi conditions, 篩選出成績大於等於60分，且為一年級的資料
df.loc[(df['成績'] >= 60) & (df['就讀班級'].str.contains('一'))].head()
```

```python
# ~ -> 取反, 在需要取反的mask前面加上這個符號(多條件要加入多次)
df.loc[~(df['成績'] >= 60) & ~(df['就讀班級'].str.contains('一', na=False))].head()
```

```python
# 用between取得符合區間數值的資料(包含)
df.loc[(df['成績'].between(80, 90))]
```
