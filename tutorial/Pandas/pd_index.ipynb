[Reference](https://www.youtube.com/watch?v=W9XjRYFkkyw&ab_channel=CoreySchafer)

```python
import pandas as pd
```

```python
df = pd.read_csv(
    '../dataset/學生修課資料與成績.csv',
    encoding='utf-8-sig',
    # index_col='學號'  # 在Read file時就設定Index
)
```

```python
df.head()
```

```python
# 設定學號為index
df.set_index('學號', inplace=True)
df.head()
```

```python
# 用index取得所有學生的學號清單
df.index
```

```python
# 用loc grab指定學生的資料
df.loc['A1607']
```

```python
# 選取欄位
df.loc['A1607', ['就讀班級', '成績']]
```

```python
df.iloc[0:2]
```

```python
# Reset Index
df.reset_index(inplace=True)
df.head()
```

```python
# sort_index, default=Ascending
df.set_index('學號', inplace=True)
df.sort_index(inplace=True, ascending=True)  # Default=True
df
```
